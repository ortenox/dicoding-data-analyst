## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
sns.set_theme(style="whitegrid")

# Add air quality categorize
def categorize_no2(value):
    if value > 50:
        return "Buruk"
    elif value >= 30:
        return "Sedang"
    else:
        return "Baik"

data_no2 = {
    'Station': ['Aotizhongxin', 'Nongzhanguan', 'Guanyuan', 'Gucheng', 'Dongsi', 'Tiantan', 'Changping', 'Shunyi', 'Huairou', 'Dingling'],
    'NO2': [59.31, 58.10, 57.90, 55.88, 53.70, 53.16, 44.19, 43.91, 32.51, 27.59]
}
df_no2 = pd.DataFrame(data_no2)

df_no2["Category"] = df_no2["NO2"].apply(categorize_no2)

# Color based on categorize
palette = {"Buruk": "red", "Sedang": "orange", "Baik": "green"}

# Create the bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="NO2", y="Station", hue="Category", dodge=False, data=df_no2, palette=palette)
plt.xlabel("NO2 Concentration")
plt.ylabel("Station")
plt.title("NO2 Levels Across Different Stations")
```

### Pertanyaan 2:

```python
data_so2 = {
    'Month': ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    'SO2': [30.42, 26.97, 26.88, 13.67, 13.47, 7.48, 5.29, 4.37, 6.02, 8.39, 15.02, 24.25]
}

df_so2 = pd.DataFrame(data_so2)

# Add air quality categorize
def categorize_so2(value):
    if value > 25:
        return "Buruk"
    elif value >= 10:
        return "Sedang"
    else:
        return "Baik"

df_so2["Category"] = df_so2["SO2"].apply(categorize_so2)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x="Month", y="SO2", hue="Category", dodge=False, data=df_so2, palette=palette)
plt.xlabel("Month")
plt.ylabel("SO2 Concentration")
plt.title("SO2 Levels Throughout the Year")
```

## Clustering

```python
import pandas as pd

# Station data
data = {
    'Station': ['Aotizhongxin', 'Nongzhanguan', 'Guanyuan', 'Gucheng', 'Dongsi',
                'Tiantan', 'Changping', 'Shunyi', 'Huairou', 'Dingling'],
    'NO2': [59.31, 58.10, 57.90, 55.88, 53.70, 53.16, 44.19, 43.91, 32.51, 27.59]
}

# Change data to datafrane
df = pd.DataFrame(data)

# Quality based on NO2
def categorize(no2):
    if no2 > 50:
        return 'Buruk'
    elif 40 <= no2 <= 50:
        return 'Sedang'
    else:
        return 'Baik'

# Menambahkan kolom kategori ke DataFrame
df['Kategori'] = df['NO2'].apply(categorize)

# Show grouping
print(df)
```
